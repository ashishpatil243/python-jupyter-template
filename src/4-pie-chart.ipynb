### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'], startangle=140)

# Title
plt.title('Programming Language Usage Distribution')

# Show plot
plt.show()
```
