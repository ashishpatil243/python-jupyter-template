### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11, 1)  # x values from 0 to 10
y = x ** 2  # y values as the square of x

# Create line plot
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='y = x²')

# Customizations
plt.title('Square Function Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis (x²)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show plot
plt.show()
```
